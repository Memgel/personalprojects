```python
import asyncio
import json
import aiohttp
from understat import Understat
!pip install nest-asyncio
import nest_asyncio
nest_asyncio.apply()
import matplotlib.pyplot as plt
```

```python
async def main():
    async with aiohttp.ClientSession() as session:
        understat = Understat(session)
        players = await understat.get_league_players(
            "epl",
            2021,
            position = "F S"
        )
        return json.dumps(players)

result = asyncio.run(main())
results_list = json.loads(result)
```

```python
results_list
```

```python
player_name = []
goals = []

for i in range(20): 
  player_name.append(results_list[i]["player_name"])
  goals.append(int(results_list[i]["goals"]))
```

```python
plt.bar(player_name,goals)
plt.xticks(rotation='vertical')
```

```python
plt.bar(player_name, goals, color='purple')
plt.xlabel("Player Names")
plt.ylabel("Goals")
plt.title("Goals scored in the 2021/2022 by Fowards")
plt.xticks(rotation='vertical')

plt.show()
```
